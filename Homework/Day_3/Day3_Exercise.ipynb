## Develop a method to rename the column names
... to "good_column_name"s (lower_case, no blanks, alpha_numeric). Hint: Use the parameter `inplace=True`

````python
# method to rename column names
def rename_columns(col_names:list) -> dict:
    col_mapper = {}
    for i in col_names:
        j = i.replace(' ', '_').lower()
        if '(' in i:
            j = j[ : i.find('(')]
        col_mapper[i] = j
    return col_mapper
````

````python
cols_mapper = rename_columns(hgnc.columns)
hgnc.rename(columns=cols_mapper, inplace=True)
````
